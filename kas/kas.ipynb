```python
from argparse import Namespace
import sys
import os
from loss import ELBOLoss
from torch.utils.data import DataLoader
sys.path.append(os.getcwd())
from data_utils import get_dataloader
# from dataset.datasets import NewRegression
import numpy as np
import random
from tqdm import tqdm
import torch

from temp_data import Temperatures
from informed_nps import INP
from plot import plot_predictive


config = Namespace(
    input_dim=1,
    output_dim=1,
    xy_encoder_num_hidden=2,
    xy_encoder_hidden_dim=128,
    data_agg_func='sum',
    latent_encoder_num_hidden=2,
    decoder_hidden_dim=64,
    decoder_num_hidden=2,
    decoder_activation='gelu',
    hidden_dim=128,
    x_transf_dim=128,
    x_encoder_num_hidden=1,
    test_num_z_samples=32,
    train_num_z_samples=1,
    knowledge_extractor_num_hidden=0,
    knowledge_extractor_hidden_dim=128,
    knowledge_dropout=0,
    knowledge_dim=128,
    knowledge_merge='self-attention',
    text_encoder = 'roberta',
    use_knowledge=False,
    freeze_llm=True,
    tune_llm_layer_norms=False,
    freeze_meta_networks=False,
    # dataset
    batch_size=64,
    min_num_context=3,
    max_num_context=20,
    x_sampler='random-uniform-25',
    noise=0,
    # reproducibility
    seed=44,
    dataset='temperatures',
    knowledge_type='desc',
    num_targets=50,
    path='latent'
    #xy_self_attention='dot',
)
config.device = "cuda"

train_dataset = Temperatures(split='train')
val_dataset = Temperatures(split='train')
test_dataset = Temperatures(split='train')
# dataset = NewRegression(root='data/new', split='train')
#dataset = CelebA(root='data', split='train', knowledge_type='set')
# dataset = metaMIMIC(config, split='train', root='data/metaMIMIC', knowledge_type='none')
# train_dataloader = get_dataloader(dataset, config)
# config.num_classes = dataset.num_classes
#dataset = SetKnowledgeTrendingSinusoids(split='train', root='data/toy-regression', knowledge_type='masked')
train_dataloader = get_dataloader(train_dataset, config)
val_dataloader = get_dataloader(val_dataset, config)
test_config = config
test_config.batch_size = 1
test_dataloader = get_dataloader(test_dataset, test_config)
# train_dataloader = DataLoader(dataset, batch_size=64, shuffle=True)
#config.knowledge_input_dim = dataset.knowledge_input_dim

model = INP(config).to(config.device)
print(model)
print(sum(p.numel() for p in model.parameters() if p.requires_grad))
loss_func = ELBOLoss()

optimizer = torch.optim.Adam(model.parameters(), lr=5e-4)

torch.manual_seed(config.seed)
np.random.seed(config.seed)
random.seed(config.seed)

epoch_losses = []
for epoch in range(1000):
    epoch_loss = 0
    num_batches = len(train_dataloader)

    for i, batch in enumerate(train_dataloader):
        optimizer.zero_grad()
        
        context, target, knowledge, _ = batch
        # print(len(knowledge))
        # print(context.shape)
        # print(knowledge.shape)
        x_context, y_context = context
        x_target, y_target = target
        x_context = x_context.to(config.device)
        y_context = y_context.to(config.device)
        x_target = x_target.to(config.device)
        y_target = y_target.to(config.device)
    
        if config.use_knowledge:
            outputs = model(x_context, y_context, x_target, y_target, knowledge)
        else:
            outputs = model(x_context, y_context, x_target, y_target, None)
    
        p_yCc = outputs[0]
    
        loss = loss_func(outputs, y_target)
        epoch_loss += loss["loss"].item()

        loss["loss"].backward()
        optimizer.step()
    epoch_losses.append(epoch_loss/num_batches)
    print(f"ep: {epoch}, avg_epoch_loss: {epoch_loss/num_batches}")

    if epoch % 10 == 0:
        with torch.no_grad():
            num_val_epochs = 1
            avg_val_epoch_loss = 0
            avg_val_nll = 0
            num_val_batches = len(val_dataloader)
            
            for _ in range(num_val_epochs):
                # model.training = False
                val_epoch_loss = 0
                val_epoch_nll = 0
                for i, batch in enumerate(val_dataloader):
                
                    context, target, knowledge, _ = batch
                    # print(len(knowledge))
                    # print(context.shape)
                    # print(knowledge.shape)
                    x_context, y_context = context
                    x_target, y_target = target
                    x_context = x_context.to(config.device)
                    y_context = y_context.to(config.device)
                    x_target = x_target.to(config.device)
                    y_target = y_target.to(config.device)
                
                    if config.use_knowledge:
                        outputs = model(x_context, y_context, x_target, y_target, knowledge)
                    else:
                        outputs = model(x_context, y_context, x_target, y_target, None)
                
                    p_yCc = outputs[0]
                
                    val_loss = loss_func(outputs, y_target)
                    val_epoch_loss += val_loss["loss"].item() / num_val_batches
                    val_epoch_nll += val_loss["log_p"].item() / num_val_batches
                avg_val_epoch_loss += val_epoch_loss / num_val_epochs
                avg_val_nll += val_epoch_nll / num_val_epochs
            print(f"ep: {epoch}, Validation Loss: {avg_val_epoch_loss}, NLL: {avg_val_nll}")
            # model.training = True
        
        # plot
        for j, batch in enumerate(test_dataloader):
            context, target, knowledge, _ = batch
            x_context, y_context = context
            x_target, y_target = target
            print(len(knowledge))
            x_context = x_context.to(config.device)
            y_context = y_context.to(config.device)
            x_target = x_target.to(config.device)
            y_target = y_target.to(config.device)
            plot_predictive(model, x_context, y_context, x_target, y_target, knowledge, config)
            break
        # probs = outputs[0].probs
        # predictions = torch.tensor(probs > 0.5, dtype=torch.float32)
        # accuracy = (predictions == y_target).float().mean()
    
        # print(accuracy)
```
